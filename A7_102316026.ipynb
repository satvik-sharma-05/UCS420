## Q1. You are a data analyst at an e-commerce company. Your task is to analyze sales data, clean and process it, perform staƟsƟcal analysis, and generate visual insights to help the business make informed decisions. 


## Part I: Randomized Sales Data GeneraƟon (NumPy)


1. IniƟalize Random Seed


2. Generate Sales Data


 Create a NumPy array of random sales data for 12 months (shape = (12, 4)),


where each column represents a product category:


 Electronics, Clothing, Home & Kitchen, Sports.


 Sales values should be between 1000 and 5000 units per month.


 The range should be randomized using the seed.


3. Convert to DataFrame


 Convert the cleaned NumPy array into a Pandas DataFrame with:


i. Columns: Electronics, Clothing, Home & Kitchen, Sports


ii. Index: Months from Jan to Dec 




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Initialize random seed (Example Roll Number: 420)
np.random.seed(420)

# Step 2: Generate sales data for 12 months (random values between 1000 and 5000)
sales_data = np.random.randint(1000, 5000, (12, 4))
    
# Step 3: Convert to DataFrame
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
categories = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]
df_sales = pd.DataFrame(sales_data, columns=categories, index=months)
print("Sales DataFrame:\n", df_sales)
```

## Part II: Data ManipulaƟon and Analysis (Pandas)
1. Display the first 5 rows and summary staƟsƟcs of the DataFrame.
2. Calculate the total sales per category and total sales per month.
3. Calculate the average sales growth between consecuƟve months for each category.
4. Add the following columns to the DataFrame:
 Total Sales: Sum of all categories per month.
 Growth Rate: Percentage change in Total Sales from the previous month.
5. Apply a randomized discount:
 If your roll number is even, apply a 10% discount to the Electronics category.
 If odd, apply a 15% discount to the Clothing category. 


```python
# Display the first 5 rows
print("First 5 rows of sales data:\n", df_sales.head())

# Summary statistics
print("\nSummary statistics:\n", df_sales.describe())

# Step 1: Total sales per category
total_category_sales = df_sales.sum()
print("\nTotal sales per category:\n", total_category_sales)

# Step 2: Total sales per month
df_sales["Total Sales"] = df_sales.sum(axis=1)
print("\nTotal sales per month:\n", df_sales["Total Sales"])

# Step 3: Average sales growth between consecutive months for each category
avg_growth = df_sales.diff().mean()
print("\nAverage sales growth between months:\n", avg_growth)

# Step 4: Growth Rate (percentage change in total sales)
df_sales["Growth Rate"] = df_sales["Total Sales"].pct_change() * 100
print("\nGrowth Rate:\n", df_sales["Growth Rate"])

# Step 5: Applying discount based on roll number
roll_number = 420  # Change this to your actual roll number

if roll_number % 2 == 0:
    df_sales["Electronics"] *= 0.90  # 10% discount if roll number is even
else:
    df_sales["Clothing"] *= 0.85  # 15% discount if roll number is odd

print("\nDiscount applied:\n", df_sales)
```

## Part III: VisualizaƟons (Matplotlib and Seaborn)
1. Plot monthly sales trends for each category using line plots.
2. Create the following plots:
 Box plots to show the sales distribuƟon for each category.


```python
# Step 1: Monthly sales trends for each category
plt.figure(figsize=(10, 5))
for category in categories:
    sns.lineplot(x=df_sales.index, y=df_sales[category], label=category)

plt.title("Monthly Sales Trends for Each Category")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.grid()
plt.show()

# Step 2: Box Plot for sales distribution
plt.figure(figsize=(8, 5))
sns.boxplot(data=df_sales[categories])
plt.title("Sales Distribution per Category")
plt.xlabel("Categories")
plt.ylabel("Sales")
plt.show()
```

## Q.2 For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])
 i. Find element-wise absolute value


ii. Find the 25th, 50th, and 75th percenƟle of flaƩened array, for each column, for each row.


iii. Mean, Median and Standard DeviaƟon of flaƩened array, of each column, and each row



```python
import numpy as np

# Step 1: Define the array
array = np.array([[1, -2, 3], [-4, 5, -6]])

# Step 2: Element-wise absolute values
abs_array = np.abs(array)
print("Absolute values:\n", abs_array)

# Step 3: Compute percentiles
percentiles_flat = np.percentile(array, [25, 50, 75])
percentiles_col = np.percentile(array, [25, 50, 75], axis=0)
percentiles_row = np.percentile(array, [25, 50, 75], axis=1)

print("\nPercentiles (Flattened Array):", percentiles_flat)
print("\nPercentiles (Column-wise):\n", percentiles_col)
print("\nPercentiles (Row-wise):\n", percentiles_row)

# Step 4: Compute mean, median, and standard deviation
mean_flat = np.mean(array)
median_flat = np.median(array)
std_flat = np.std(array)

mean_col = np.mean(array, axis=0)
median_col = np.median(array, axis=0)
std_col = np.std(array, axis=0)

mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)

print("\nMean (Flattened):", mean_flat)
print("Median (Flattened):", median_flat)
print("Standard Deviation (Flattened):", std_flat)
print("\nMean (Column-wise):", mean_col)
print("Median (Column-wise):", median_col)
print("Standard Deviation (Column-wise):", std_col)
print("\nMean (Row-wise):", mean_row)
print("Median (Row-wise):", median_row)
print("Standard Deviation (Row-wise):", std_row)
```

## Q.3 For the array: a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0]).
 i. Find floor, ceiling and truncated value, rounded values 


```python
import numpy as np

# Step 1: Define the array
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

# Step 2: Compute floor, ceiling, truncation, and rounded values
floor_values = np.floor(a)
ceiling_values = np.ceil(a)
truncated_values = np.trunc(a)
rounded_values = np.round(a)

print("\nFloor Values:", floor_values)
print("Ceiling Values:", ceiling_values)
print("Truncated Values:", truncated_values)
print("Rounded Values:", rounded_values)
```

## Q.4 WAP to swap two elements in a list in Python Using a Temporary Variable. 


```python
# Step 1: Define the list
lst = [10, 20, 30, 40, 50]

# Step 2: Swap elements (Example: swap index 1 and index 3)
temp = lst[1]
lst[1] = lst[3]
lst[3] = temp

print("\nList after swapping:", lst)
```

## Q.5 Write a Python program to swap two elements in a set by converƟng it to a list first.

```python
# Step 1: Define the set
s = {10, 20, 30, 40, 50}

# Step 2: Convert to list and swap elements
s_list = list(s)
s_list[1], s_list[3] = s_list[3], s_list[1]  # Swap index 1 and index 3

# Convert back to set
s_swapped = set(s_list)
print("\nSet after swapping:", s_swapped)
```
